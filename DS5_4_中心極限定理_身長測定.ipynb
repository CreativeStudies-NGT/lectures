### 5.4 中心極限定理のデモ ― 身長の繰り返し測定

真値 μ = 160 cm、測定誤差 σ = 1 cm の身長を繰り返し測定するシミュレーション。  
**N 回の測定の平均値**は標準誤差 σ/√N の正規分布に従う（中心極限定理）。  
測定回数を増やすほど、平均値の分布が真値の周囲に集中していくことを可視化する。

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import japanize_matplotlib
import ipywidgets as widgets

%config InlineBackend.figure_format = 'retina'

TRUE_HEIGHT = 160  # 真値 (cm)
SIGMA = 1          # 測定誤差の標準偏差 (cm)
N_REPEAT = 2000    # 標本平均を計算する繰り返し回数
```

```python
# 1回の測定の分布
measurements = np.random.normal(TRUE_HEIGHT, SIGMA, N_REPEAT)
x = np.linspace(TRUE_HEIGHT - 4*SIGMA, TRUE_HEIGHT + 4*SIGMA, 500)

plt.figure(figsize=(8, 4))
plt.hist(measurements, bins=40, density=True, facecolor='steelblue', ec='k', alpha=0.7,
         label=f'測定値 ({N_REPEAT}回)')
plt.plot(x, norm.pdf(x, TRUE_HEIGHT, SIGMA), 'r-', lw=2,
         label=f'理論値 N(μ={TRUE_HEIGHT}, σ={SIGMA})')
plt.axvline(TRUE_HEIGHT, color='k', ls='--', lw=1.5, label=f'真値 {TRUE_HEIGHT} cm')
plt.xlabel('測定値 (cm)', fontsize=13)
plt.ylabel('確率密度', fontsize=13)
plt.title('1回の身長測定の分布', fontsize=15)
plt.legend()
plt.tight_layout()
```

```python
# 測定回数 N を変えたときの標本平均の分布を比較
N_values = [1, 4, 16, 100]
colors = ['royalblue', 'seagreen', 'orange', 'crimson']
x = np.linspace(TRUE_HEIGHT - 3.5, TRUE_HEIGHT + 3.5, 500)

plt.figure(figsize=(9, 5))
for N, color in zip(N_values, colors):
    samples = np.random.normal(TRUE_HEIGHT, SIGMA, (N_REPEAT, N))
    means = samples.mean(axis=1)
    se = SIGMA / np.sqrt(N)
    plt.hist(means, bins=60, density=True, alpha=0.35, color=color, ec='none')
    plt.plot(x, norm.pdf(x, TRUE_HEIGHT, se), color=color, lw=2,
             label=f'N={N:3d}  σ/√N = {se:.3f} cm')

plt.axvline(TRUE_HEIGHT, color='k', ls='--', lw=1.5, label=f'真値 {TRUE_HEIGHT} cm')
plt.xlabel('N 回測定の平均値 (cm)', fontsize=13)
plt.ylabel('確率密度', fontsize=13)
plt.title('測定回数 N による標本平均分布の変化', fontsize=15)
plt.xlim(TRUE_HEIGHT - 3.5, TRUE_HEIGHT + 3.5)
plt.legend(fontsize=11)
plt.tight_layout()
```

```python
def plot_mean_dist(N):
    samples = np.random.normal(TRUE_HEIGHT, SIGMA, (N_REPEAT, N))
    means = samples.mean(axis=1)
    se = SIGMA / np.sqrt(N)
    x = np.linspace(TRUE_HEIGHT - 3.5, TRUE_HEIGHT + 3.5, 500)

    plt.figure(figsize=(8, 4))
    plt.hist(means, bins=60, density=True, facecolor='steelblue', ec='k', alpha=0.6,
             label=f'標本平均の分布 (N={N}, {N_REPEAT}回)')
    plt.plot(x, norm.pdf(x, TRUE_HEIGHT, se), 'r-', lw=2,
             label=f'理論値: σ/√N = {se:.3f} cm')
    plt.axvline(TRUE_HEIGHT, color='k', ls='--', lw=1.5, label=f'真値 {TRUE_HEIGHT} cm')
    plt.xlabel('N 回測定の平均値 (cm)', fontsize=13)
    plt.ylabel('確率密度', fontsize=13)
    plt.title(f'N = {N} 回測定の標本平均分布  (標準誤差 = {se:.3f} cm)', fontsize=14)
    plt.xlim(TRUE_HEIGHT - 3.5, TRUE_HEIGHT + 3.5)
    plt.legend(fontsize=11)
    plt.tight_layout()
    plt.show()

slider = widgets.IntSlider(
    value=1, min=1, max=200, step=1,
    description='N (測定回数):',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='500px')
)
widgets.interact(plot_mean_dist, N=slider)
```
